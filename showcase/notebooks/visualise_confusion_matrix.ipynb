```python
import numpy as np
from sklearn.metrics import ConfusionMatrixDisplay
```

Define relevant functions

```python
def check_available_points_on_curve(dir: str, file_name: str):
    path = "../../" + dir + file_name
    stats = np.load(path)
    return stats["points_on_curve"]


def load_predictions(set, point_on_curve, dir: str, file_name: str):
    path = "../../" + dir + file_name
    stats = np.load(path)
    id_point_on_curve = np.isclose(stats["points_on_curve"], point_on_curve)
    targets = stats[set + "_targets"][id_point_on_curve][0]
    predictions = stats[set + "_predictions"][id_point_on_curve][0]
    return targets, predictions
```

 Set directory and file

```python
dir = "results/MNIST_CNNBN/evaluation_curve/"

file_name_model_init = "predictions_and_probabilities_curve_epoch0.npz"

file_name_model_trained = "predictions_and_probabilities_curve_epoch15.npz"
```

 Check which points of curve have been evaluated

```python
print(check_available_points_on_curve(dir, file_name_model_init))

print(check_available_points_on_curve(dir, file_name_model_trained))
```

 Set evaluation parameters

```python
set = "train"

point_on_curve = 0.5
```

 Plot confusion matrix for curve before training.

```python
targets, predictions = load_predictions(set, point_on_curve, dir, file_name_model_init)

ConfusionMatrixDisplay.from_predictions(targets, predictions)
```

 Plot confusion matrix for curve after training.

```python
targets, predictions = load_predictions(set, point_on_curve, dir, file_name_model_trained)

ConfusionMatrixDisplay.from_predictions(targets, predictions)
```
